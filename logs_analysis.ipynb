```python
import pandas as pd
import json
from datetime import datetime

# Cargar JSON de logs
try:
    with open('data/logs.json','r',encoding='utf-8') as fh:
        logs = json.load(fh)
    df = pd.DataFrame(logs)
    df['Fecha'] = pd.to_datetime(df['timestamp'])
    print('Loaded', len(df), 'logs')
except FileNotFoundError:
    print('data/logs.json not found. If you exported CSV, load it via pd.read_csv.')

# Ejemplo: latencia por d√≠a
if 'df' in globals() and not df.empty:
    df['total_time'] = df['metrics'].apply(lambda x: x.get('total_time',0.0))
    print(df.groupby(df['Fecha'].dt.date)['total_time'].mean())
```
